```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv.imread("drive/MyDrive/CV_data/SIFT1_img.jpg")
img2 = cv.imread("drive/MyDrive/CV_data/SIFT2_img.jpg")
gray= cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img)
cv2_imshow(img2)
```

```python
def runSIFT(image1, image2):
  sift = cv.SIFT_create()
  # Detect and compute key points
  kp1, des1 = sift.detectAndCompute(image1,None)
  kp2, des2 = sift.detectAndCompute(image2, None)
  # Draw key points on the image
  image1 = cv.drawKeypoints(image1, kp1, image1, flags= cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
  cv2_imshow(image1)
  image2 = cv.drawKeypoints(image2, kp2, image2, flags= cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
  cv2_imshow(image2)
  bf = cv.BFMatcher()
  # Match 2 keypoints on 2 images
  matches = bf.match(des1, des2)
  matches = sorted(matches, key = lambda x:x.distance)
  img3 = cv.drawMatches(image1,kp1,image2,kp2,matches[:int(len(matches)*10/100)],None,flags=cv.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)
  cv2_imshow(img3)
```

```python
runSIFT(img,img2)
```
